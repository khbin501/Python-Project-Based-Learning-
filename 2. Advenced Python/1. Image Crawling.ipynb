```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 1. Service 객체를 생성하여 드라이버 경로를 전달합니다.
service = Service(ChromeDriverManager().install())

# 2. Chrome 드라이버를 실행할 때 service 인자를 명시합니다.
driver = webdriver.Chrome(service=service)

URL = 'https://www.google.com/imghp'
driver.get(url=URL)

driver.implicitly_wait(time_to_wait=10)
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By

elem = driver.find_element(By.CSS_SELECTOR,"#APjFqb")
elem.send_keys("아이리 칸나")
elem.send_keys(Keys.RETURN)
```

```python
links=[]
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
driver.implicitly_wait(5)
images = driver.find_elements(By.CSS_SELECTOR,".YQ4gaf")

for image in images:
    if image.get_attribute('src') is not None:
        links.append(image.get_attribute('src'))
        
print("찾은 이미지 개수 :",len(links))
```

```python
import urllib.request

for i,k in enumerate(links):
    url = k
    urllib.request.urlretrieve(url,"C:/Users/ha050/Desktop/Python-Project-Based-Learning-/Crawling/"+str(i+1)+".jpg")
    
print("크롤링 완료")
```
